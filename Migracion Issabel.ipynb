### <font color = 'Blue'> Troncales Punto2 a Nuevo conmutador

```python
connection = netmiko.ConnectHandler(ip='10.7.175.2', device_type= 'linux', username= 'root', port='22', password='cecomelastixdos2017')
sip_add=connection.send_command(" cat /etc/asterisk/sip_additional.conf")
iax_add=connection.send_command(" cat /etc/asterisk/iax_additional.conf")
connection.disconnect()
sip_add+='\n'
iax_add+='\n'
print(sip_add)
print("**********************************************************************")
print(iax_add)
```

```python
#SIP
conmutadorNvo_trunk=[]
pattern=re.compile(r'\[(?!\d{4})(\w+)\](.*?)\n\n', re.DOTALL) 
matches=pattern.findall(sip_add)
length=len(list(matches))
for index,match in enumerate(matches):
    dict_punto2=next((trunk_dict for trunk_dict in punto_2_trunk.json()['results'] if trunk_dict['trunk_name']==match[0]), None)
    if dict_punto2==None:
        print(f"No se encuentra la troncal {match[0]} en punto_2_trunk (via RestAPI)")
        continue
    peer_json=json.loads('{"'+match[1][1:].replace('=','":"').replace('\n','",\n"')+'"}')
    dict_punto2['channel_name']=dict_punto2['name']
    dict_punto2['peer']=peer_json
    conmutadorNvo_trunk.append(dict_punto2)
    nuevo_conmutador=requests.post(dir_nuevo_conmutador+'/trunks/', verify=False, headers= {"content-type": "application/json","Authorization": "Bearer " + token2['access_token']}, data=json.dumps(dict_punto2))
    print(str(index)+" de "+str(length)+nuevo_conmutador.text+" OK")
#     if index==2:   #RAPIDO PARA PRUEBAS
#         break
    
#IAX
pattern=re.compile(r'\[(?!\d{4})(\w+)\](.*?)\n\n', re.DOTALL) 
matches=pattern.findall(iax_add)
length=len(list(matches))
for index,match in enumerate(matches):
    dict_punto2=next((trunk_dict for trunk_dict in punto_2_trunk.json()['results'] if trunk_dict['trunk_name']==match[0]), None)
    if dict_punto2==None:
        print(f"No se encuentra la troncal {match[0]} en punto_2_trunk (via RestAPI)")
        continue
    peer_json=json.loads('{"'+match[1][1:].replace('=','":"').replace('\n','",\n"')+'"}')
    dict_punto2['channel_name']=dict_punto2['name']
    dict_punto2['technology']=dict_punto2['tech']
    dict_punto2['peer']=peer_json
    conmutadorNvo_trunk.append(dict_punto2)
    nuevo_conmutador=requests.post(dir_nuevo_conmutador+'/trunks/', verify=False, headers= {"content-type": "application/json","Authorization": "Bearer " + token2['access_token']}, data=json.dumps(dict_punto2))
    print(str(index)+" de "+str(length)+nuevo_conmutador.text+" OK")
#     if index==2:  #RAPIDO PARA PRUEBAS
#         break
    
    
```
